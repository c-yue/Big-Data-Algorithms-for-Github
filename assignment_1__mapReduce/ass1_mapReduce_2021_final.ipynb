```python
# Here a simple function that reads a .csv file, removes the header,
# and writes the output on a .txt file having the same name of the input file + the keyword "no_header"
# input: a csv file
def remove_header(filename) :
  with open(filename + ".csv",'r') as f:
    with open(filename + "no_header.txt",'w') as f1:
        next(f) # skip header line
        for line in f:
            f1.write(line)
  return f1

remove_header('small_file')
print("file correctly saved")
```

### <strong> Exercise 1 - Almost-empty flights</strong> 
#### <strong> 4 points </strong>
Describe and define a MapReduce procedure that gives as output the number of flights that departed with 10% of capacity maximum (i.e. empty, 1%, 6%, etc.) . 

The output can be of any form you like. 

You can use any data structure you want to support your implementation.




```python
# import library
from collections import defaultdict
import numpy as np
import pandas as pd # only used for read data
```

```python
# read flight data and transform every row to a list
data_tb = pd.read_csv("medium_file.csv")
data_list = data_tb.values.tolist()
```

```python
# map 
# mapping logic: devide flights into 3 set by
# calculating the capital level of every flight and comparing with 10%
def map_e_flights(list_2):
    #to be completed
    cap_lvl = 0.1
    map_res = []
    for list_1 in list_2:
      psg = float(list_1[4])
      seat = float(list_1[5])
      if seat == 0:
        t = ('miss_info',1)   
      elif psg/seat > cap_lvl:
        t = ('high_cap',1)
      else:
        t = ('low_cap',1)
      map_res.append(t)
    return map_res

# shuffle
# put flight with the same level of capital value in the same shuffle
def shuffle_e_flights(map_list):
    #to be completed
    shuffle = defaultdict(list)
    for k,v in map_list:
      shuffle[k].append(v)
    return shuffle

# reduce
# group the flight number of different capital value
def reduce_e_flights(shuffle_dict):
    #to be completed
    res = []
    for key in shuffle_dict:
      res.append((key,sum(shuffle_dict[key])))
    return res
```

```python
# operate the process
reduce_e_flights(shuffle_e_flights(map_e_flights(data_list)))
```

```python
# map
# deviding data by month and
# mapping according to the connection of the flight
def map_top_connections(list_2):
    #to be completed
    map_res = defaultdict(list)
    for list_1 in list_2:
      month = datetime.datetime.strptime(list_1[8],"%Y-%m-%d").strftime("%Y-%m")
      connection = ' <==> '.join((set(list_1[2:4]))) # Special expression for connection made by Yue CHEN
      connection_value = (connection, 1)
      map_res[month].append(connection_value)
    return map_res

# shuffle
# put flight with the same connection in the same shuffle
def shuffle_top_connections(map_dict):
    #to be completed
    shuffle_dict = defaultdict()
    for month in map_dict:
      shuffle = defaultdict(list)
      for k,v in map_dict[month]:
        shuffle[k].append(v)
      shuffle_dict[month] = shuffle
    return shuffle_dict

# reduce
# group the flight number of different connection
def reduce_top_connections(shuffle_dict):
    #to be completed
    res_dict = defaultdict()
    for month in shuffle_dict:
      res = []
      for key in shuffle_dict[month]:
        res.append((key,sum(shuffle_dict[month][key])))
      res.sort(key = lambda x: x[1], reverse = True)
      res_dict[month] = (res[:5])
    return res_dict
```

```python
# mapreduce using data_list, vrb and logic
# input data, variables to count and calculate logic, so that the mapreduce process can be used with different devide logic
# here the type of date, variables to count and calculate logic is list, int, function

# map
# mapping according to a logic, which is a function that can be defined later
def map_with_logic(list_2, vrb, logic):
    #to be completed
    map_res = []
    for list_1 in list_2:
      if logic(vrb, list_1):
        map_res.append((True,1))
      else:
        map_res.append((False,1))
    return map_res

# shuffle
# put flight with the same logic in the same shuffle
def shuffle_with_logic(map_list):
    #to be completed
    shuffle = defaultdict(list)
    for k,v in map_list:
      shuffle[k].append(v)
    return shuffle

# reduce
# group the flight number of different logic
def reduce_with_logic(shuffle_dict):
    #to be completed
    res = []
    for key in shuffle_dict:
      res.append((key,sum(shuffle_dict[key])))
    res.sort(key = lambda x: x[1], reverse = True)
    return res
```

```python
# operate the process

# define the column to be cal
vrb = 4

# define the logic function to devide flights
# test if a flight is full
def logic(col_num, flight_list):
  res = False
  pasg = flight_list[col_num]
  seat = flight_list[col_num+1]
  if pasg >= seat:
    res = True
  return res

# mapreduce using data_list, vrb and logic function
reduce_with_logic(shuffle_with_logic(map_with_logic(data_list, vrb, logic)))
```
